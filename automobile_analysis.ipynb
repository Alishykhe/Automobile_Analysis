EDA ANALYSIS

```python
com = sns.countplot(data= df1, x = 'company')
sns.countplot(rc={'figure.figsize':(15,5)})
plt.xticks(rotation=90)
for bars in com.containers:
    com.bar_label(bars)
```

```python
plt.figure(figsize= (12,6))
sns.boxplot(data= df1, x= 'company', y= 'price' , palette = 'coolwarm', hue= 'company')
plt.xticks(rotation=90)
plt.title('Price Trends Across Car Brands')
plt.show()
```

Luxury brands like BMW, Mercedes-Benz, and Porsche have the highest price variations, while budget brands like Chevrolet, Dodge, and Nissan show consistent pricing. Outliers in Mercedes and Porsche indicate ultra-luxury models.

```python
grouped = df1.groupby('engine')['price'].mean()
sns.barplot( data= df1, x = 'engine' , y= 'price', hue = 'company')
plt.xlabel('Engine type')
plt.ylabel('Average price')
plt.title('Average price by engine type')
```

```python
sns.scatterplot(x= 'Mileage', y= 'engine', data = df1, hue = 'company')
```

Different car brands use a variety of engine types but OHC engines are the most common across different mileage levels. DOHC engines tend to be in lower-mileage cars, while rotor engines are quite rare showing they are used in specific cases.

```python
sns.countplot(data = df1, x= 'Style', palette = 'Set2', hue = 'Style', legend = False)
plt.title("Most Popular Car Styles in the Market")
plt.xticks(rotation=45)
plt.show()
```

The sedan is the most popular car style in the market, followed by hatchbacks while convertibles and hardtops are the least common.

CONCLUSION:

This analysis highlights that most fuel-efficient cars have OHC engines, while DOHC engines offer a mix of mileage and performance. Additionally, sedans dominate the market, proving that buyers prioritize comfort and practicality over specialized styles.
